Vamos a hacer distintos graficos importando los videos de las filmaciones, y que un programa "openpiv" nos haga graficos de campos de velocidades frame por frame del video importado. Es decir que va a graficar entre dos frames consecutivos para donde se esta moviendo el fluido que esta rotando. 

Importamos las librerias

```python
import numpy as np
import matplotlib.pyplot as plt 
from openpiv import tools, pyprocess, validation, filters, scaling 
from openpiv import pyprocess, piv
import cv2 
```

Importamos el video y graficamos para donde se mueve el campo de velocidades. 

```python
vidcap = cv2.VideoCapture('vel_hiperlenta_70frames.avi') # acá puede ir cualquiera de los videos de las filmaciones
success, image1 = vidcap.read()
count = 0
U = []
V = []

while success and count < 10:
    # cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
    success, image2 = vidcap.read()
    # print('Read a new frame: ', success)
    if success:
        x,y,u,v = piv.simple_piv(image1.sum(axis=2), image2.sum(axis=2),plot=True);
        image1 = image2.copy()
        count += 1
        U.append(u)
        V.append(v)
```
